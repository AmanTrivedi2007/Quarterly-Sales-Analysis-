```python

import pandas as pd
encs = ["utf-8", "utf-8-sig", "cp1252", "latin1", "iso-8859-1", "utf-16", "utf-32", "ascii"]
for enc in encs:
    try:
        df = pd.read_csv(r'C:\Users\AMAN TRIVEDI\Desktop\Python codes\data analysis\sales data\sales_data_sample.csv',encoding = enc)
        break     
    except UnicodeDecodeError:
        continue
df.drop_duplicates()


grouped = df.groupby('QTR_ID')['SALES'].sum()
print(f'Total slaes per qtr is \n{grouped}')

grouped_per_month = df.groupby('MONTH_ID')['SALES'].sum()
print(f'Total sales done per month is \n {grouped_per_month}')
print("\n")
total_sales = df['SALES'].sum()
print(f'Total sales till now: \n{total_sales}$')
print("\n")
avearage_order_value = df['SALES'].mean()
print(f'The average order value is \n{avearage_order_value}')





```

```python
country_distribution =  df.groupby('COUNTRY')['SALES'].sum()
print(f'Total sales per country \n {country_distribution}')

```

```python
product_sales_distribution =  df.groupby('PRODUCTCODE')['SALES'].sum()
print(f'Total sales done as per product code \n {product_sales_distribution}')
print("\n")
product_sales_distribution2 = df.groupby('PRODUCTLINE')['SALES'].sum()
print(f'Total sales done PER PRODUCT \n {product_sales_distribution2}')
print("\n")
top_product = product_sales_distribution2.idxmax()
top_product_slaes = product_sales_distribution2.max()

print(f"The most product sold is - {top_product} - with a sales of -  {top_product_slaes}$")
```

```python

product_sales_distribution_deal_size = df.groupby('DEALSIZE')['SALES'].sum()
print(f'Total sales as per deal size \n {product_sales_distribution_deal_size}')

import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.bar(product_sales_distribution2.index,product_sales_distribution2.values)
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product vs sales graph')
plt.show()



```

```python
import matplotlib.pyplot as plt
s = product_sales_distribution_deal_size  # your Series
plt.pie(s.values, labels=s.index, autopct="%.1f%%"); plt.axis("equal"); 
plt.title('Sales Distribution as per Deal Size')
plt.show()
```

```python
Status_info = df.groupby('STATUS')['QUANTITYORDERED'].sum()
print(Status_info)

plt.pie(Status_info.values, labels=Status_info.index ,autopct="%.1f%%"); plt.axis("equal"); 
plt.title('Status of the order quantity in the phases of delivary')
plt.show()
```
